```python
import numpy as np      
import matplotlib.colors                                
import matplotlib.pyplot as plt 
from scipy.signal import find_peaks
import scipy
import matplotlib as mpl
import scipy.io
from matplotlib.gridspec import GridSpec
from Functions_for___TheoreticalModel import *
```

```python
# Parameters
vi = 0.01
Vol = vi*10**(-14); NA = 6.02*10**(23); Cal = NA*Vol*10**(-6);
TC = 0.158
kc1 = 0.15*Cal*TC;                      # production of p53
kc2 = 0.1*TC;                           # degradation of p53 by mdm2     
kc3 = 0.1*Cal;                          # degradation of p53 by mdm2
kc4 = 0.1/Cal*TC;                       # production of mdm2 - mRNA
kc5 = 0.1*TC                            # degradation of mdm2 - mRNA
kc6 = 0.2*TC                            # production of mdm2
kc7 = 0.1*TC                            # degradation of mdm2
kc8 = 0.0036                            # binding of mdm2 and nutlin !!!!!! THIS IS PROBABLY SMALLER LIKE 0.0001
dt = 1
```

```python
def make_p53_plot(ax1,ax2,t,x,Text,nut_const,tstart_nut,colorp53,color_right,alpha):
    ax1.plot(t/60,x[:,3],label = "p53", color = colorp53, alpha = alpha)
    ax1.plot(t/60,x[:,0], label ="nutlin", color = "tab:gray",alpha = alpha)
    ax1.set_xlabel("Time [hour]")
    ax1.set_title(f"Text: {Text/(60)} h, Aext: {nut_const}")
    #ax1.legend()
    tstart_index = int(len(t)/2)
    ax2.plot(x[tstart_index:,2],x[tstart_index:,1],x[tstart_index:,3], color = color_right, alpha = alpha)
    ax2.set_ylabel("Mdm2 mRNA")
    ax2.set_xlabel("Mdm2 free")
    #
    plt.tight_layout()
```

```python
def make_simulation_and_plot(Text, nut_const,tend_nut, dt, simT, f_p53,x0,
                             kc1,kc2,kc3,kc4,kc5,kc6,kc7,kc8,colorp53,color_right, alpha, tstart_nut):                                       
    t = np.arange(0, simT, dt)       
    x = odeRungeKutta4_p53(f_p53, x0, dt, simT, tstart_nut,tend_nut, 
                           nut_const,Text, 
                           args = (kc1,kc2,kc3,kc4,kc5,kc6,kc7,kc8,))
    make_p53_plot(ax1,ax2,t,x,Text,nut_const,tstart_nut,colorp53,color_right,alpha)
```

```python
x0ivec = [50]
tstart_nut = 0#3*5.5*60 # Start of natlin pulse [min]

Text_vec = np.array([11])*60 #np.linspace(0.1,3,1)*5.5*60
nut_const_vec = np.array([1.5])#([0,0.5,1.0,1.5,1.9,2.0,2.1,10,100])#np.linspace(0.001,170,1)

for Text_ith,Text in enumerate(Text_vec):          # iterate over nutlin periods
    simT = 10*Text # Time of simulation equal to 100 p53 periods[min] 
    
    tend_nut = tstart_nut+Text/2 # End of natlin pulse after 40 mins [min] 
    make_simulation_and_plot(5.5*60, 0, 0, dt, simT, f_p53, x0,
                         kc1,kc2,kc3,kc4,kc5,kc6,kc7,kc8, "black", "black", 1, tstart_nut)
    
    for nut_ith,nut_const  in enumerate(nut_const_vec):    # iterate over nutlin concentrations
        fig = plt.figure(figsize=(10, 2.5))
        gs = GridSpec(1, 4, figure=fig)
        # Define the subplots using the grid layout
        ax1 = fig.add_subplot(gs[:, :3])  # Spans 3 columns
        ax2 = fig.add_subplot(gs[:, 3],projection='3d')   # Spans 1 column
        for x0i in x0ivec:
            x0 = [0,1,50,x0i]   #Initial condition

            alpha = 1/(nut_ith+1)
            make_simulation_and_plot(Text, nut_const, tend_nut, dt, simT, f_p53,x0,
                                 kc1,kc2,kc3,kc4,kc5,kc6,kc7,kc8, "tab:green", "tab:green", 1, tstart_nut)
            ax2.set_zlabel("p53")
        plt.show()
```

```python
def autocorr(x):
    n = x.size
    norm = (x - np.mean(x))
    result = np.correlate(norm, norm, mode='same')
    acorr = result[n//2 + 1:] / (x.var() * np.arange(n-1, n//2, -1))
    #lag = np.abs(acorr).argmax() + 1
    time = np.arange(0,len(acorr))
    peaks = find_peaks(acorr)[0]
    lag = time[np.where(acorr == np.max(acorr[peaks]))]
    r = acorr[lag-1]        
    if np.abs(r) > 0.5:
      print('Appears to be autocorrelated with r = {}, lag = {}'. format(r, lag))
    else: 
      print('Appears to be not autocorrelated')
    return r, lag, acorr
```

```python
x = odeRungeKutta4_p53(f_p53,x0,dt,simT, tstart_nut,tend_nut, nut_const,Text, args = (kc1,kc2,kc3,kc4,kc5,kc6,kc7,kc8,))
signal = x[:,3]
r,lag,acorr = autocorr(signal)
plt.figure(figsize = (5,2.5))
plt.ylabel("Autocorrelation")
plt.plot(acorr)
plt.show()
plt.figure(figsize=(12,2))
time = np.arange(0,len(signal))
plt.ylabel("Signal")
plt.plot(time,signal, label = "signal")
plt.plot(time+lag,signal, label = "signal+lag")
plt.legend()
plt.show()
```
